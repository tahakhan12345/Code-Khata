```python
def function_cell_8(user_input):
    import requests
    import json

    # Function to query the ThreatFox API
    def query_threatfox(ip_address, api_key=None):
        url = 'https://threatfox-api.abuse.ch/api/v1/'

        # Define the request body in JSON format
        data = {
            'query': 'search_ioc',
            'search_term': ip_address
        }

        headers = {
            'Content-Type': 'application/json'  # Set the content type to JSON
        }

        # If an API key is provided, include it in the headers
        if api_key:
            headers['API-Key'] = api_key

        try:
            # Send the POST request to the ThreatFox API with JSON data
            response = requests.post(url, json=data, headers=headers)
            response.raise_for_status()  # Check for HTTP errors

            # Get the JSON response
            result = response.json()

            # Print the result to the terminal
            print("API Response:")
            print(json.dumps(result, indent=4))

            # Check if 'query_status' is 'no_result'
            if result.get('query_status') == 'no_result':
                print("No results found for the given IP address.")
            # Ensure 'data' is present and it's a list if there are results
            elif 'data' in result and isinstance(result['data'], list):
                if len(result['data']) > 0:
                    for entry in result['data']:
                        # Ensure 'entry' is a dictionary before accessing keys
                        if isinstance(entry, dict):
                            print(f"Indicator: {entry.get('ioc', 'N/A')}")
                            print(f"Threat type: {entry.get('threat_type', 'N/A')}")
                            print(f"Malware: {entry.get('malware', 'N/A')}")
                            print(f"Confidence: {entry.get('confidence', 'N/A')}")
                            print(f"First seen: {entry.get('first_seen', 'N/A')}")
                            print(f"Last seen: {entry.get('last_seen', 'N/A')}")
                            print(f"Tags: {entry.get('tags', 'N/A')}")
                            print(f"Reporter: {entry.get('reporter', 'N/A')}")
                            print("-" * 40)
                else:
                    print("No data found for the given IP address.")
            else:
                print("'data' key missing or not a list in the API response.")

        except requests.exceptions.RequestException as e:
            print(f"Error querying ThreatFox API: {e}")

    # Main function to execute the query for a static IP
    def main():
        # Define a static IP address to query
        static_ip = "185.220.101.2 "  # Replace with the desired static IP address

        # Your ThreatFox API key (if needed)
        api_key = "ebab174dfef4aac484f4e66e6be47499"  # Replace with your actual API key, or leave as None if no key is needed

        # Query ThreatFox with the static IP address and API key
        query_threatfox(static_ip, api_key)

    # Run the main function
    if __name__ == "__main__":
        main()



    return locals()
```
